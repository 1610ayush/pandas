```python
food1 = pd.read_csv("indian_food.csv")
```

```python
food2 = pd.read_csv("indianFood.csv")
```

```python
# INNER JOIN - return rows which have matching values in both table
food_merge = food1.merge(food2, on="name") #merging two data frames, on argument takes the similar column of both data frames and merge dem on basis of specified column
```

```python
food_merge.head()
```

```python
food_merge.shape
```

```python
food_merge = food1.merge(food2, on="name", suffixes= ("1","2")) #changing suffixes x and y by 1 and 2 respectively
```

```python
food_merge
```

```python
#merging one the basis ofmore than one column
food1.merge(food2, on=["name","ingredients"])
```
