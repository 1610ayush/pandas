# PIVOT TABLES

```python
data.pivot_table(values="Units Sold",index="Item Type") #values specify the column which we want to summarize and index specify the data items of column which we group by. And by default pivot table returns the mean of each group
```

```python
import numpy as np
data.pivot_table(values="Units Sold",index="Item Type",aggfunc=np.median) #we can change the default value by aggfunc
```

```python
data.pivot_table(values="Units Sold",index="Item Type",columns="Sales Channel") #to group by two variables we specify the 2nd variable in columns 
```

```python
data.pivot_table(values="Units Sold",index="Item Type",columns="Sales Channel",fill_value=0,margins=True) #fill_value fills the missing values with its specified value and margins returns one row and column which contains the mean of dat respective row and column by leaving missing values
```

```python
data.isna().sum() #tells no of missing values in each column
```
